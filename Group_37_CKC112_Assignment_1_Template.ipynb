**The Traveler-Centric Approach.**

Treat the Traveler as the primary entity. Each Traveler object "owns" the hotels and reviews they interact with.

* Person.h (Base)

* Traveler.h (Inherits Person)

* Review.h (Inside Traveler via Composition)

```python
%%writefile Constants.h
#ifndef Constants_H
#define Constants_H

const int MAX_HOTELS = 50;
const int MAX_TRAVELERS = 100;
const int MAX_REVIEWS = 200;

#endif
```

```python
%%writefile Review.h
#ifndef Review_H
#define Review_H

#include <iostream>
#include <string>
#include<iomanip>

using namespace std;

class Review {
private:
    int reviewID;
    int userID;
    string hotelName;
    double rating;
    string reviewText;

public:
    Review() : reviewID(0), userID(0), hotelName(""), rating(0.0), reviewText("") {}

    Review(int rid, int uid, string hotel, double r, string text)
        : reviewID(rid), userID(uid), hotelName(hotel), rating(r), reviewText(text) {}


    int getReviewID() const { return reviewID; }
    int getUserID() const { return userID; }
    string getHotelName() const { return hotelName; }
    double getRating() const { return rating; }
    string getReviewText() const { return reviewText; }

    bool operator>(const Review& other) const {
        return this->rating > other.rating;
    }

    bool operator<(const Review& other) const {
        return this->rating < other.rating;
    }

    void print() const {
        cout << "  Review ID: " << reviewID << " | User: " << userID
             << " | Rating: " << fixed << setprecision(1) << rating << "/5.0" << endl;
        cout << "  Text: " << reviewText << endl;
    }

};

#endif
```

```python
%%writefile Person.h
#ifndef Person_H
#define Person_H

#include<iostream>
#include<string>
#include "Constants.h"
using namespace std;

class Person
{
  protected:
    int id;
    string name;
    string email;

  public:
    Person(): id(0), name(""), email(""){}
    Person(int pId, string pN, string pE): id(pId), name(pN), email(pE){}

    // Getter functions
    int getPersonID() const {return id;}
    string getPersonName() {return name;}
    string getPersonEmail() {return email;}

    // Setter functions
    void setPersonName(string n) { name = n; }
    void setPersonEmail(string e) {email = e; }

    virtual void printProfile() const
    {
      cout  << "ID: " << id
            << " | Name: " << name
            << " | Email: " << email << endl;
    }

    virtual ~Person(){}
};
#endif
```

```python
%%writefile Traveler.h
#ifndef Traveler_H
#define Traveler_H

#include<iostream>
#include "Person.h"
#include "Review.h"
#include "Constants.h"
using namespace std;

class Traveler: public Person
{
  private:
    string membershipLevel;
    int loyaltyPoints;
    Review writtenReview[MAX_REVIEWS];
    int reviewCount;

  public:
    Traveler(): Person(), membershipLevel(""), loyaltyPoints(0), reviewCount(0){}

    // Default constrcutor
    Traveler(int pId, string pN, string pE, string ml, int lp) :
    Person(pId, pN, pE), membershipLevel(ml), loyaltyPoints(lp), reviewCount(0){}

    // Constructor
    Traveler(int pId, string pN, string pE, string ml, int lp, Review* r, int rc):
    Person(pId, pN, pE), membershipLevel(ml), loyaltyPoints(lp), reviewCount(rc){}

    // Getter functions
    string getMembershipLevel() const {return membershipLevel;}
    int getLoyaltyPoints() const {return loyaltyPoints;}
    int getReviewCount() const { return reviewCount; } // Added getter for reviewCount

    // Setter functions
    void setMembershipLevel(string ml) {membershipLevel = ml;}
    void setLoyaltyPoints(int points) {loyaltyPoints = points;}

    bool addReview(const Review& review)
    {
      if (reviewCount >= MAX_REVIEWS) return false;

      writtenReview[reviewCount++] = review;

      // Awarding loyalty points for writing review
      int pointsAwarded = static_cast<int>(review.getRating()) * 10;
      loyaltyPoints += pointsAwarded;
      updateMembershipLevel();

      return true;
    }

    Review* getAllReviews()
    {
      return writtenReview;
    }

    void updateMembershipLevel()
    {
      if (loyaltyPoints >= 10000) membershipLevel = "Platinum"; // Adjusted thresholds
      else if (loyaltyPoints >= 5000) membershipLevel = "Gold";
      else if (loyaltyPoints >= 1000) membershipLevel = "Silver";
      else membershipLevel = "Bronze";
    }

    void printProfile() const override // Changed name to match base class
    {
        cout << "========== TRAVELER PROFILE ==========" << endl;
        Person::printProfile();
        cout << "Membership Level: " << membershipLevel << endl;
        cout << "Loyalty Points: " << loyaltyPoints << endl;
        cout << "Reviews Written: " << reviewCount << endl;
        cout << "=====================================" << endl;
    }

     void displayWrittenReviews() const {
        if (reviewCount == 0) {
            cout << "This traveler has not written any reviews." << endl;
            return;
        }
        cout << "\nAuthor: " << name << " - Reviews (" << reviewCount << "):" << endl;
        for (int i = 0; i < reviewCount; i++) {
            cout << "[" << (i+1) << "] ";
            writtenReview[i].print();
        }
    }

    // Compare travelers by loyalty points
    bool operator>(const Traveler& right) const {
        return this->loyaltyPoints > right.loyaltyPoints;
    }

    // Added operator< for sorting
    bool operator<(const Traveler& right) const {
        return this->loyaltyPoints < right.loyaltyPoints;
    }
};

#endif
```

```python
%%writefile Hotel.h
#ifndef Hotel_H
#define Hotel_H

#include<iostream>
#include<string>
#include<iomanip>
#include "Review.h"
#include "Constants.h"
using namespace std;

class Hotel
{
  private:
    int hotelID;
    string hotelName;
    string location;
    Review hotelReview[MAX_REVIEWS];
    int reviewCount;
    double totalRating;

  public:
    Hotel() : hotelID(0), hotelName(""), location(""), reviewCount(0), totalRating(0.0){}

    Hotel(int hid, string hName, string hloca, int rc, double t) :
    hotelID(hid), hotelName(hName), location(hloca), reviewCount(rc), totalRating(t){}

    // Constructor for creating new hotels when loading reviews
    Hotel(int hid, string hName, string hloca) :
    hotelID(hid), hotelName(hName), location(hloca), reviewCount(0), totalRating(0.0){}

    // Getter functions
    int getHotelID() const {return hotelID;} // Fixed syntax
    string getHotelName() const {return hotelName;}
    string getLocation() const {return location;}
    int getReviewCount() const {return reviewCount;}
    double getAverageRating() const
    {
      if (reviewCount == 0) return 0.0;
      else return totalRating / reviewCount;
    }

    // Add review to hotel's collection
    bool addReview(const Review& review) {
        if (reviewCount >= MAX_REVIEWS) return false;

        hotelReview[reviewCount++] = review;
        totalRating += review.getRating();
        return true;
    }

    // Get review via index
    Review getReview(int index) const
    {
      if (index >= 0 && index < reviewCount) return hotelReview[index];
      else return Review();
    }

    Review* getAllReviews()
    {
      return hotelReview;
    }

    // Display hotel information and all reviews
    void printHotelProfile() const {
        cout << "\n========== HOTEL: " << hotelName << " ==========" << endl;
        cout << "ID: " << hotelID << " | Location: " << location << endl;
        cout << "Total Reviews: " << reviewCount << endl;
        cout << "Average Rating: " << fixed << setprecision(2) << getAverageRating() << "/5.0" << endl;
        cout << "======================================\n" << endl;

        if (reviewCount == 0) {
            cout << "No reviews available for this hotel." << endl;
            return;
        }

        cout << "REVIEWS:" << endl;
        for (int i = 0; i < reviewCount; i++) {
            cout << "[" << (i+1) << "] ";
            hotelReview[i].print(); // Fixed typo: hotelReviews to hotelReview
        }
    }

    // Operator overloading - comparing hotels by ratings
    bool operator>(const Hotel& right) const
    {
      return this->getAverageRating() > right.getAverageRating();
    }

    bool operator<(const Hotel& right) const
    {
      return this->getAverageRating() < right.getAverageRating();
    }

    // Overload the << operator to print Hotel objects (for topPickMatcher)
    friend ostream& operator<<(ostream& os, const Hotel& hotel) {
        os << hotel.getHotelName() << " (Avg Rating: "
           << fixed << setprecision(2) << hotel.getAverageRating() << ")";
        return os;
    }
};

#endif
```

```python
%%writefile FileIO.h
#ifndef FileIO_H
#define FileIO_H

#include <iostream>
#include <fstream>
#include <string>
#include <iomanip>
#include "Traveler.h"
#include "Hotel.h"
using namespace std;

inline void loadTravelers(Traveler travelers[], int& travelerCount) {
    ifstream file("users.txt");
    if (!file) {
        cout << "Warning: users.txt not found." << endl;
        travelerCount = 0;
        return;
    }

    travelerCount = 0;
    char u;
    int id, points;
    string name, location, email, level;
    char ch;

    while (travelerCount < MAX_TRAVELERS && file >> u >> id) {
      file.get(ch);

      getline(file, name, '\t');
      getline(file, location, '\t');
      getline(file, email, '\t');
      getline(file, level, '\t');
      file >> points;

      Traveler traveler(id, name, email, level, points);
      travelers[travelerCount++] = traveler;
    }

    file.close();
    cout << "Loaded " << travelerCount << " travelers." << endl;
}

inline void loadReviews(Traveler travelers[], int travelerCount, Hotel hotels[], int& hotelCount) {
    ifstream file("reviews.txt");
    if (!file) {
        cout << "Warning: reviews.txt not found." << endl;
        return;
    }

    hotelCount = 0;
    int reviewIDCounter = 1;

    char u;
    int userID;
    double rating;
    string reviewText, hotelName;
    char ch;

    while (file >> u >> userID >> rating) {
      file.get(ch);

      getline(file, reviewText, '\t');
      getline(file, hotelName);

      Review rev(userID, (int)rating, reviewText, hotelName);

      for (int i = 0; i < travelerCount; i++) {
              if (travelers[i].getPersonID() == userID) {
                  travelers[i].addReview(rev);
                  break;
              }
          }

          int hotelIndex = -1;
          for (int i = 0; i < hotelCount; i++) {
              if (hotels[i].getHotelName() == hotelName) {
                  hotelIndex = i;
                  break;
              }
          }

          if (hotelIndex == -1) {
              if (hotelCount < MAX_HOTELS) {
                  hotels[hotelCount] = Hotel(hotelCount + 1, hotelName, "");
                  hotelIndex = hotelCount++;
              }
          }

          if (hotelIndex != -1) {
              hotels[hotelIndex].addReview(rev);
          }
      }

      file.close();
      cout << "Loaded reviews and created " << hotelCount << " hotels." << endl;
  }


#endif
```

```python
%%writefile Feature.h
#ifndef Feature_H
#define Feature_H

#include <iostream>
#include <fstream>
#include <string>
#include <iomanip>

#include "Constants.h"
#include "Review.h"
#include "Person.h"
#include "Traveler.h"
#include "Hotel.h"
#include "FileIO.h"
using namespace std;

void printMenu();

void hotelAnalyticsReport(const Hotel& hotel)
{
  cout << "\n--- Analytics Report for " << hotel.getHotelName() << " ---" << endl;
  cout << "(Detailed analytics would go here)" << endl;
}


void propertyDeepDive(Hotel hotels[], int hotelCount) {
    cout << "\n========== PROPERTY DEEP-DIVE ==========" << endl;
    cout << "Enter hotel name: ";
    cin.ignore();
    string searchName;
    getline(cin, searchName);

    for (int i = 0; i < hotelCount; i++) {
        if (hotels[i].getHotelName() == searchName) {
            hotels[i].printHotelProfile();
            hotelAnalyticsReport(hotels[i]);
            return;
        }
    }

    cout << "Hotel '" << searchName << "' not found in database." << endl;
}


void travelerProfileManagement(Traveler travelers[], int travelerCount) {
    cout << "\n========== TRAVELER PROFILE ==========" << endl;
    cout << "Enter traveler ID: ";
    int searchID;
    cin >> searchID;

    for (int i = 0; i < travelerCount; i++) {
        if (travelers[i].getPersonID() == searchID) {
            travelers[i].printProfile();
            travelers[i].displayWrittenReviews();
            return;
        }
    }

    cout << "Traveler with ID " << searchID << " not found." << endl;
}


void topPickMatcher(Hotel hotels[], int hotelCount) {
    cout << "\n========== TOP-PICK MATCHER ==========" << endl;
    cout << "Hotel Rankings (by average rating):" << endl;

    // Simple bubble sort using operator overloading
    Hotel temp;
    for (int i = 0; i < hotelCount - 1; i++) {
        for (int j = 0; j < hotelCount - i - 1; j++) {
            if (hotels[j] < hotels[j + 1]) {
                temp = hotels[j];
                hotels[j] = hotels[j + 1];
                hotels[j + 1] = temp;
            }
        }
    }

    cout << "\nRanked Hotels:" << endl;
    for (int i = 0; i < hotelCount && i < 10; i++) {
        cout << "[" << (i + 1) << "] " << hotels[i] << endl;
    }
}


void topReviewers(Traveler travelers[], int travelerCount) {
    cout << "\n========== TOP REVIEWERS (by Loyalty Points) ==========" << endl;

    // Simple bubble sort using operator overloading
    Traveler temp;
    for (int i = 0; i < travelerCount - 1; i++) {
        for (int j = 0; j < travelerCount - i - 1; j++) {
            if (travelers[j] < travelers[j + 1]) {
                temp = travelers[j];
                travelers[j] = travelers[j + 1];
                travelers[j + 1] = temp;
            }
        }
    }

    cout << "\nTop Reviewers:" << endl;
    for (int i = 0; i < travelerCount && i < 10; i++) {
        cout << "[" << (i + 1) << "] " << travelers[i].getPersonName()
             << " | Loyalty Points: " << travelers[i].getLoyaltyPoints()
             << " | Level: " << travelers[i].getMembershipLevel()
             << " | Reviews: " << travelers[i].getReviewCount() << endl;
    }
}

void printMenu() {
    cout << "\n========== HOTEL REVIEW SYSTEM ==========" << endl;
    cout << "1. Property Deep-Dive" << endl;
    cout << "2. Traveler Profile" << endl;
    cout << "3. Top-Pick Matcher" << endl;
    cout << "4. Top Reviewers" << endl;
    cout << "5. Exit" << endl;
    cout << "========================================" << endl;
}

#endif
```

```python
%%writefile main.cpp

#include <iostream>
#include <fstream>
#include <string>
#include <iomanip>

#include "Constants.h"
#include "Review.h"
#include "Person.h"
#include "Traveler.h"
#include "Hotel.h"
#include "FileIO.h"
#include "Feature.h"

int main() {
  cout << "========== HOTEL REVIEW SYSTEM ==========" << endl;
  cout << "Loading data..." << endl << endl;

  //Entity-centric Lifestyle
  Traveler travelers[MAX_TRAVELERS];
  int travelerCount = 0;

  Hotel hotels[MAX_HOTELS];
  int hotelCount = 0;

  // Read both files exactly once only
  loadTravelers(travelers, travelerCount);
  loadReviews(travelers, travelerCount, hotels, hotelCount);

  int choice = 0;
  while (choice != 5)
  {
    printMenu();
    cin >> choice;
    switch (choice)
    {
      case 1:
        propertyDeepDive(hotels, hotelCount);
        break;
      case 2:
        travelerProfileManagement(travelers, travelerCount);
        break;
      case 3:
        topPickMatcher(hotels, hotelCount);
        break;
      case 4:
        topReviewers(travelers, travelerCount);
        break;
      case 5:
        cout << "\nExiting system..." << endl;
        break;
      default:
        cout << "Invalid choice!" << endl;
    }
  }

  cout << "Saving data..." << endl;

  cout << "Program terminated." << endl;

  return 0;

}
```
